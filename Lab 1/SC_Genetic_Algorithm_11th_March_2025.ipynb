```python
import random
import numpy

POPULATION = 4 #even
SIZE = 5
ITERATION = 5
def initialize_population(): #pop = int, samp = int
    accepted_range = [0,1]
    popln = []
    smpl = []
    for i in range(POPULATION):
        for j in range(SIZE):
            smpl.append(random.choice(accepted_range))
        popln.append(smpl)
        smpl = []
    return popln


def fittness(sample):
    fit = []
    sum = 0
    for i in range(POPULATION):
        for j in range(SIZE):
            sum += 2**j * sample[i][j]
        fit.append(sum)
        sum = 0
    return fit

def crossover(parents): #single point crossover
    crossover_point = []
    crossover_child = []
    for i in range(POPULATION//2):
        crossover_point.append(random.randint(0,SIZE-1))
    for i in range(0,POPULATION,2):
        chld1 = parents[i]
        chld2 = parents[i+1]
        for j in range(crossover_point[i//2],SIZE):
            chld1[j], chld2[j] = chld2[j], chld1[j]
        crossover_child.append(chld1)
        crossover_child.append(chld2)
    return crossover_child, crossover_point


def expected_count(fittness):
    mean = sum(fittness)/ float(len(fittness))
    count = []
    real_count = []
    for i in range(POPULATION):
        count.append(round(fittness[i]//mean))
        real_count.append(fittness[i]/mean)
    if sum(count) > POPULATION:
        for i in range(POPULATION):
            if ((real_count[i] - count[i]) >0.4 and (real_count[i] - count[i]) < 0.6) and count[i]!=0:
                count[i] = count[i] - 1
                break
    elif sum(count) < POPULATION:
        for i in range(POPULATION):
            if ((real_count[i] - count[i]) >0.4 and (real_count[i] - count[i]) < 0.6) and count[i]!=0:
                count[i] = count[i] + 1
                break
    elif sum(count) == POPULATION:
        print("No adjustment needed")
    return  count

def select_parents(parents, count):
    selected_parents = [[] for i in range(POPULATION)]
    for i in range(POPULATION):
        for j in range(count[i]):
            while True:
                pos = random.randint(0,POPULATION-1)
                if selected_parents[pos] == []:
                    selected_parents[pos] = parents[i]
                for par in selected_parents:
                    if par == []:
                        break
                else:
                    break
    return selected_parents

popln = initialize_population()
for i in range(ITERATION):
    fitnes = fittness(popln)
    expt_cnt = expected_count(fitnes)
    select_par = select_parents(popln,expt_cnt)
    children, cross = crossover(select_par)

    print("ITERATION - ",i)
    print(f"|    Individuals    |    fittness    |    expected count    |")
    for i in range(POPULATION):
        print(f"|{str(popln[i]):^19}|{str(fitnes[i]):^16}|{str(expt_cnt[i]):^22}|")

    print("Crossover points are-  ", cross)

    print(f"|    selected parents    |      children      |")
    for i in range(POPULATION):
        print(f"|{str(select_par[i]):^24}|{str(children[i]):^20}|")
    popln = children
```
